```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.metrics import mean_squared_error 
```

```python
dataframe = datasets.load_diabetes(return_X_y=False)
```

```python
X = dataframe.data[:,:2]
Y = dataframe.target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.30, random_state=42)
```

```python
train_size = X_train.shape[0]
test_size  = X_test.shape[0]
```

```python
K_Value = 5
predictions = np.zeros((test_size,1))
```

```python
for i in range(test_size) :
    test_data = X_test[i,:]
    distances = np.zeros((train_size, 1))
    distance_list = []
    
    for j in range(train_size) :
        train_data = X_train[j,:]
        #get euclidean distance between test_data and train_data
        distances[j] = np.sqrt(np.sum((test_data - train_data)**2))
        distance_list.append((distances[j], Y_train[j]))
     
    distance_list = sorted(distance_list)
    # Find distances array find the K nearest neighbors   
    # Get the average from nearest neighbors class value
    # Fill the prediction array with individual predicted data stored in result
    # Find the loss function value
    # Find the difference from each predicted data and Y test data and square that difference
    # and add those differences. 
    # Divide the total sum of differenes with length of Y test.
    
    val = 0
    for k in range(K_Value) :
        val += 1/distance_list[k][1]
    
    
    #print(distance_list[0:K_Value])
    #nearest neighbors
    neighbors = distance_list[0:K_Value]
    print(neighbors)
    
    #result
    result= val/K_Value
    print(result)
    
    #predictions
    predictions[i] = result
```

```python
# find error
total_loss=0
m=(len(Y_test))

for i in range(len(predictions)) :
    loss = ((predictions[i] - Y_test[i])*(predictions[i] - Y_test[i]))
    total_loss += loss 

ans = total_loss/m
print(ans)
    
```
